### 2. 請試著在程式中印出以下三個 NdArray 的屬性並且解釋結果？（屬性：ndim、shape、size、dtype、itemsize、length、type）

```
a = np.random.randint(10, size=6)

```

````python
# 記得先 Import 正確的套件
import numpy as np
a = np.random.randint(10, size=6) 
b = np.random.randint(10, size=(3,4)) 
c = np.random.randint(10, size=(2,3,2)) 
'''
開始!
'''
````

````python
a = np.random.randint(10, size=6) 
print(a)
print(a.ndim)
print(a.shape)
print(a.size)
print(a.dtype)
print(a.itemsize)
print(len(a))
print(type(a))

'''a是一維'''
````

````python
b = np.random.randint(10, size=(3,4)) 
print(b)

print(b.ndim)
print(b.shape)
print(b.size)
print(b.dtype)
print(b.itemsize)
print(len(b))
print(type(b))

'''
b是二維
'''
````

````python
c = np.random.randint(10, size=(2,3,2)) 
print(c)
print(c.ndim)
print(c.shape)
print(c.size)
print(c.dtype)
print(c.itemsize)
print(len(c))
print(type(c))

'''
c是三維
'''
````

### 3. 如何利用 list(...) 實現 a.tolist() 的效果？試著用程式實作。


````python
a = np.random.randint(10, size=6) 

print(a.tolist())
print(list(a))
````

````python
b = np.random.randint(10, size=(3,4)) 

print(b.tolist())
print(list(b))
````

````python
c = np.random.randint(10, size=(2,3,2)) 

print(c.tolist())
print(list(c))
````

````python
def tolist(iterable):
    print(iterable)
    print([list(f) for f in list(iterable)])
print(tolist(a))
print(tolist(b))
print(tolist(c))
````
